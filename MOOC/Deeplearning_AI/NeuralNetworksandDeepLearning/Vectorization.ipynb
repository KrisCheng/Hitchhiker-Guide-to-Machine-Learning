```python
import time

a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a,b)
toc = time.time()
print("Vectorized version: " + str(1000*(toc-tic)) + " ms")

c = 0
tic = time.time()
for i in range(1000000):
    c += a[i]*b[i]
toc = time.time()
print("For loop version: " + str(1000*(toc-tic)) + " ms")
```

```python
a = np.random.randn(3,4)
b = np.random.randn(4,1)
c = np.zeros([3,4])
# for loop
for i in range(3):
  for j in range(4):
    c[i][j] = a[i][j] + b[j]
print(c)

# vectorization
c = np.zeros([3,4])
c = a + b.T
print(c)
```
